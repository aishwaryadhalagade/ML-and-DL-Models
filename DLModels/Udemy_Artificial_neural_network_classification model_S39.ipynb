# Artificial Neural Network

### Importing the libraries

```
import numpy as np
import pandas as pd
import tensorflow as tf
```

```
tf.__version__
```

## Part 1 - Data Preprocessing

### Importing the dataset

```
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

```
print(X)
```

```
print(y)
```

### Encoding categorical data

Label Encoding the "Gender" column

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,2] = le.fit_transform(X[:,2])
```

```
print(X)
```

One Hot Encoding the "Geography" column

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[1])],remainder= 'passthrough')
X = np.array(ct.fit_transform(X))
```

```
print(X)
```

### Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size= 0.2, random_state = 0)
```

### Feature Scaling

```
#Feature scaling is very much compulsory in deep learning to all the independent variables

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
print(X_train)
```

```
print(X_test)
```

## Part 2 - Building the ANN

### Initializing the ANN

```
# ANN model is nothing but the sequence of the layers connected togather from input,hidden upto output layer
# keras library is integrated in tensorflow
# ann(artificial neural network) is created as instance of sequential class from keras.models
# these just initialises the ANN model

ann = tf.keras.models.Sequential()
```

### Adding the input layer and the first hidden layer

```
# dense class add fully connected layers into ANN(so fully connected layer we are about to build will be the object of dense class)
# units = 6(number of hidden layer units in first hidden layer of ANN ); activation function for hidden layers are generally rectifier (relu) function

ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

### Adding the second hidden layer

```
# if hidden layer in ANN is 1=shallow type of ANN ; but more than 1 hidden layer is present then it becomes deep ANN,hence add second hidden layer
# change the values of hyperparameters to get different result(may be more accuarate)

ann.add(tf.keras.layers.Dense(units=6 , activation='relu'))
```

### Adding the output layer

```
# here we need only 1 output layer unit(neuron) as we get binary output here; but if we are gonna get classification outputs like for 3 classes seperate then we require 3 output neurons(unita)

ann.add(tf.keras.layers.Dense(units = 1, activation= 'sigmoid'))
```

## Part 3 - Training the ANN

### Compiling the ANN

```
# compile the ANN with optimizer,loss function and metric(accuracy)
# for stochastic gradient descent which reduces the loss error, we use adam optimizer
# for binary classification(0 or 1) problem use only loss= 'binary_crossentropy'
# for other than binary classification like for categorical classification(3 or more than 3 classes) use loss ='categorical_crossentropy' and also at the output layer the activation function for categorical classification is 'softmax' rather than sigmoid

ann.compile(optimizer = 'adam', loss= 'binary_crossentropy', metrics= ['accuracy'] )
```

### Training the ANN on the Training set

```
#batch size takes certain number of obeservations, and these parameter gives the number of predictions in the batch to be compared to that same number of real results(at same time)
# take batch size = 32 (generally)
# neural network has to be trained over certain number of epochs to improve the accuracy over time

ann.fit(X_train,y_train, batch_size = 32, epochs= 100 )
```

### Predicting the Test set results

```
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)                     # if y_pred(probablity) >0.5= true ie (1) else we get false ie (0)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

### Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```
